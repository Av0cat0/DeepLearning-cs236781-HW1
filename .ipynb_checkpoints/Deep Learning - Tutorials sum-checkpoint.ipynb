```python
import torch
# Tensors can be moved to GPU memory (if a GPU is available on the machine).
t = torch.tensor([[1, 2, 3], [4, 5, 6]], dtype=torch.float32)
if torch.cuda.is_available():
    t = t.cuda()
    
# Better way to use GPU tensors (and later - models)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
t = t.to(device)
t
```

```python
# Tell pytorch to track operations w.r.t. this tensor
w = torch.tensor([1, 2, 3], dtype=torch.float32, requires_grad=True)
# Do some calculations with w
y = torch.exp(-w)
z = y ** 3 - 2
# Aggregate so that we get a scalar
l = torch.sum(z)
l
```

```python
from torch import autograd

# This is where the magic happens
w_grad = autograd.grad(l, w)
w_grad
```

Each element i
of w_grad is actually the derivative of l with respect to that element in w!

$\left[\mathtt{w_{grad}}\right]_i=\frac{\partial \ell}{\partial w_i}$

Lets look at the computation graph resulting from the above calculations.

```python
import torchviz
torchviz.make_dot(l, params=dict(w=w))
```
